### Lets categorize the states in 4 parts on the basis of gdp per capita

```python
gsva_states[gsva_states[['Per Capita GSDP (Rs.)']]<gsva_states['Per Capita GSDP (Rs.)'].quantile(.2)]['Per Capita GSDP (Rs.)']
```

```python
C4 = list(gsva_states[gsva_states['Per Capita GSDP (Rs.)']<=gsva_states['Per Capita GSDP (Rs.)'].quantile(.2)].index)
```

```python
C3 = list(gsva_states[gsva_states['Per Capita GSDP (Rs.)']<=gsva_states['Per Capita GSDP (Rs.)'].quantile(.5)].index & gsva_states[gsva_states['Per Capita GSDP (Rs.)']>gsva_states['Per Capita GSDP (Rs.)'].quantile(.2)].index)
```

```python
C2 = list(gsva_states[gsva_states['Per Capita GSDP (Rs.)']<=gsva_states['Per Capita GSDP (Rs.)'].quantile(.85)].index & gsva_states[gsva_states['Per Capita GSDP (Rs.)']>gsva_states['Per Capita GSDP (Rs.)'].quantile(.5)].index)
```

```python
C1 = list(gsva_states[gsva_states['Per Capita GSDP (Rs.)']<=gsva_states['Per Capita GSDP (Rs.)'].quantile(1)].index & gsva_states[gsva_states['Per Capita GSDP (Rs.)']>gsva_states['Per Capita GSDP (Rs.)'].quantile(.85)].index)
```

```python
len(C1), len(C2), len(C3), len(C4)
```

```python
gsva_states.shape
```

```python
subsector = ['Agriculture, forestry and fishing', 'Crops', 'Livestock',
       'Forestry and logging', 'Fishing and aquaculture',
       'Mining and quarrying','Manufacturing',
       'Electricity, gas, water supply & other utility services',
       'Construction','Trade, repair, hotels and restaurants',
       'Trade & repair services', 'Hotels & restaurants',
       'Transport, storage, communication & services related to broadcasting',
       'Railways', 'Road transport', 'Water transport', 'Air transport',
       'Services incidental to transport', 'Storage',
       'Communication & services related to broadcasting',
       'Financial services',
       'Real estate, ownership of dwelling & professional services',
       'Public administration', 'Other services']
```

```python
gsva_states.loc[C1,subsector]
```

```python
top_sector_C1 = pd.DataFrame()
top_sector_C2 = pd.DataFrame()
top_sector_C3 = pd.DataFrame()
top_sector_C4 = pd.DataFrame()
for i in C1:
    top_sector_C1=top_sector_C1.append(gsva_states.loc[[i],subsector].T.sort_values(i,ascending=False).head(3))
for i in C2:
    top_sector_C2=top_sector_C2.append(gsva_states.loc[[i],subsector].T.sort_values(i,ascending=False).head(3))
for i in C3:
    top_sector_C3=top_sector_C3.append(gsva_states.loc[[i],subsector].T.sort_values(i,ascending=False).head(3))
for i in C4:
    top_sector_C4=top_sector_C4.append(gsva_states.loc[[i],subsector].T.sort_values(i,ascending=False).head(3))
```

```python
fig,ax=plt.subplots(1,2,figsize=[20,7])
top_sector_C1.plot.bar(ax=ax[0])
ax[0].set_title('C1')
ax[0].legend(loc=2)
top_sector_C2.plot.bar(ax=ax[1])
ax[1].set_title('C2')
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=[20,7])
top_sector_C3.plot.bar(ax=ax[0], legend='left')
ax[0].set_title('C3')
top_sector_C4.plot.bar(ax=ax[1])
ax[1].set_title('C4')
plt.show()
```

Inferences-
1. States with highest gdp per capita is more inclined towards secondary(manufacturing) and tertiary sectors(real-estate) except haryana.
2. States with low gdp per capita is more inclined towards primary sector mainly agricuture.
3. Haryana is the only state with high gdp per capita where highest contribution is from agriculture. Even though haryana have highest contribution from tertiary sector but still it have highest contribution from agriculture in subcategory.
